```
# part 1
import pandas as pd
import numpy as np
df = pd.read_csv("https://gist.githubusercontent.com/tiangechen/b68782efa49a16edaf07dc2cdaa855ea/raw/0c794a9717f18b094eabab2cd6a6b9a226903577/movies.csv")
df = df.drop_duplicates()
def find_top_worldwide_gross():
  data_columns = df.columns
  numpy_array = df.to_numpy()
   

  top_worldwide_gross_array = [data_columns]
  for item in numpy_array:
    if float(item[6].split("$")[1]) > 350:
      top_worldwide_gross_array.append(item)
  top_worldwide_gross_numpy_array = np.array(top_worldwide_gross_array)
  new_df = pd.DataFrame(top_worldwide_gross_numpy_array, columns=top_worldwide_gross_numpy_array[0]) 
  new_df = new_df.drop([0])
  return new_df

top_worldwide_gross_df = find_top_worldwide_gross()

top_worldwide_gross_df[top_worldwide_gross_df["Audience score %"] > 70]
```

```
# part 1
import pandas as pd
import numpy as np
df = pd.read_csv("https://gist.githubusercontent.com/tiangechen/b68782efa49a16edaf07dc2cdaa855ea/raw/0c794a9717f18b094eabab2cd6a6b9a226903577/movies.csv")
df = df.drop_duplicates()
def convert_to_float():
  float_dictionary = {}
  numpy_array = df.to_numpy()
  for item in numpy_array:
    float_dictionary[item[6]] = float(item[6].split("$")[1])
  return float_dictionary
df["Worldwide Gross"].replace(convert_to_float(), inplace=True)

df[(df["Audience score %"] > 70) & (df["Worldwide Gross"] > 350)]

```

```

import pandas as pd
import numpy as np
df = pd.read_csv("https://gist.githubusercontent.com/tiangechen/b68782efa49a16edaf07dc2cdaa855ea/raw/0c794a9717f18b094eabab2cd6a6b9a226903577/movies.csv")
df = df.drop_duplicates()
# part 1

df["Worldwide Gross"] = pd.to_numeric(df["Worldwide Gross"].str.replace("$", ""))
df[(df["Audience score %"] > 70) & (df["Worldwide Gross"] > 350)]

```
